```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load file as strings
data = np.genfromtxt(
    "ASTR19_F25_group_project_data.txt",
    dtype=str,
    comments="#"
)

# Extract columns
day = data[:,0].astype(float)
time_str = data[:,1]
height = data[:,2].astype(float)

# Convert HH:MM â†’ decimal hours
def time_to_hour(t):
    h, m = t.split(":")
    return int(h) + int(m)/60

hour = np.array([time_to_hour(t) for t in time_str])

# Convert to elapsed hours
elapsed_hours = (day - 1)*24 + hour

# Quick preview
for i in range(5):
    print(day[i], time_str[i], hour[i], height[i])
```

```python
def tide_model(t, A0, A1, phi1, P1, A2, phi2, P2):
    return (A0
            + A1 * np.cos(2*np.pi*t/P1 + phi1)
            + A2 * np.cos(2*np.pi*t/P2 + phi2))
```

```python
np.savetxt(
    "processed_data.txt",
    np.column_stack([elapsed_hours, height]),
    header="elapsed_hours height"
)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
resid = np.loadtxt("residuals.txt")
resid_std = np.std(resid, ddof=1)

print("Residual Std Dev:", resid_std)
```

```python
tsunami_increase = 2.0
z = tsunami_increase / resid_std
print(f"2 ft deviation = {z:.2f} sigma")
```

```python
resid_out = np.append(resid, tsunami_increase)
```

```python
q25, q75 = np.percentile(resid_out, [25,75])
iqr = q75 - q25
n = len(resid_out)

bin_width = 2 * iqr / (n ** (1/3))
bins = int((resid_out.max()-resid_out.min())/bin_width)

plt.figure(figsize=(6,4))
plt.hist(resid_out, bins=bins)
plt.xlabel("Residual (ft)")
plt.ylabel("Count")
plt.title("Residuals with 2-ft Outlier Added")
plt.grid(True)
plt.savefig("residuals_hist_with_outlier.pdf")
plt.show()
```
